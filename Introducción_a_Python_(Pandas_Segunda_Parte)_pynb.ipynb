# ¿Cómo funciona el tipo de cambio?

Cuando nos referimos al tipo de cambio hablamos de **el valor de una divisa respecto de otra divisa**

Por ejemplo, el precio del peso mexicano suele referenciarse al dólar estadounidense.

Con datos al 04 de marzo:

$$1 USD = 17.43 MXN$$

Esto es, **para comprar 1 dólar estadounidense necesitaríamos 17.43 pesos mexicanos.**

Una forma equivalente de pensarlo, es que:

$$1 MXN = 0.05737 USD$$

Que de forma análoga, nos indica que para comprar 1 peso mexincano necesitaríamos 1/17.45 = 0.05737 dólares, menos de 6 centésismas de dólar.

Así, la equivalencia sería:

$$x USD/MXN = 1/x MXN/USD$$

Esta forma de comparar el valor de una divisa repecto de otra se conoce como **tipo de cambio.**



En el mercado cambiario se han establecido convenciones para dar precio a una divisa respecto de otra. Naturalmente, para conocer el desempeño de una divisa respecto de otra solemos fijar respecto a qué divisa queremos conocer su valor.

Por ejemplo, la convención (notación):

$$17.45 USD/MXN$$

Indica que necesitaríamos 17.45 pesos ($MXN$) para comprar un dólar americano ($USD$).

Por su parte, tenemos:

$$0.92 USD/EUR$$

Nos indica que necesitamos 0.92 euros (EUR) para comprar 1 dólar (USD).

Mientras que:

$$18.45 EUR/MXN $$

Indica que el valor de 1 EUR es de 18.45 MXN.

La referencia del precio de una divisa suele referenciarse respecto del dólar para conocer su desempeño.

Decimos que **una divisa se aprecia respecto de otra** si el tipo de cambio baja. Por ejemplo, si el tipo de cambio USD/MXN pasa de 17.45 a 17.30:

$$17.45USD/MXN ---> 17.30 USD/MNX $$

El peso se apreció 15 centavos (MXN) respecto del dólar. Equivalentemente, esto nos diría que el dólar se depreció 15 centavos (MXN).

Equivalentemente, si el tipo de cambio pasa:

$$17.45 USD/MXN ----> 18.00 USD/MXN$$

El peso se habría depreciado 15 centavos (MXN) respecto del dólar, mientras que el dólar se habría apreciado 15 centavos (MXN) respecto del peso.

## ¿Cómo podemos medir el desempeño del dólar americano (USD)?

El mercado cambiario a nivel mundial tiene la convención de referenciar el precio de su divisa respecto del USD. Las razones principales son:

*   Estados Unidos es la economía más grande e importante del mundo
*   La divisa que más se opera (compra y vende) en el mundo es el USD
*   El país que mayor liquidez provee al mercado (es decir, demanda y oferta productos) es Estados Unidos

Así que para el año de 1971, la Reserva Federal (Banco Central de Estados Unidos) creó el índice del dólar, conocido como $USDX$, o popularmente conocido como $DXY$.

Este índice, similar a otros índices accionarios como el SP500, permite conocer el desempeño del dólar al compararlo con otras divisas.

* Saber con exactitud si el dólar se está apreciaando o depreciando respecto al resto de divisas nos llevaría a contrastar el dólar con todas y cada una de las divisas en el mundo.

* No obstante, y como es normal en el mercado, esta tarea se simplifica a medir el desempeño del dólar únicamente contra aquellas divisas cuyo peso sea significativo.



DXY composition.png

La fórmula para calcular el valor actual del $DXY$ es la siguiente:

$$DXY = 50.14348112 × EURUSD^{-0.576} × USDJPY^{0.136} × GBPUSD^ {-0.119} × USDCAD^{0.091} × USDSEK^{0.042} × USDCHF^{0.036} $$

*Donde cada uno de los exponentes representa el peso de dicha divisa en el índice (ver tabla)*

<br><font color = #3F6E9A size=5> NOTA </font>

*Cuando el USD que se fija como 1, el exponente del tipo de cambio es positivo.*

*Cuando la divisa foránea se fija como 1, el exponente es negativo.*

**Este índice está en unidades y toma en cuenta como 100.00 unidades el año base (1973). Desde entonces, solo ha tenido un rebalanceo en 1990.**

Es decir, si la lectura hoy del DXY es que se ubica en **120 puntos**, nos está indicando que el USD se está **apreciando** un 20%.

Por otro lado, si la lectura del índice se ubica en **80 puntos**, nos está indicando que el USD se está **depreciando** un 20%,

# Cálculo del P&L

Tenemos dos convenciones, una en la que las pérdidas son positivas, y otra en que las pérdidas son negativas.

**CONVENCIÓN DEL CURSO**
* Cuando las pérdidas son negativas, estamos en una serie de **Ganancias y Pérdidas (L&P)**
* Cuando las pérdidas son positivas, estamos en una serie de **Pérdidas y Gancias (P&L)**

```python
dxy_serie['Loss & Profit'] = dxy_serie['Close'] - dxy_serie['Close'].shift(1)
dxy_serie
```

```python
dxy_serie['P&L'] = dxy_serie['Close'].diff().multiply(-1)
dxy_serie
```

Distribution on the basis of skewness value:

* $Skewness = 0$: Then normally distributed.
* $Skewness > 0$: Then more weight in the left tail of the distribution.
* $Skewness < 0$: Then more weight in the right tail of the distribution.

Distribution on the basis of kurtosis value:

* $Kurtosis = 3$: Then normal distribution.
* $Kurtosis > 3$: It signifies that it tries to produce more outliers rather than the normal distribution. It is called leptokurtic
* $Kurtosis < 3$ : It is called playkurtic

```python
kur = [dxy_serie['P&L'].kurtosis(), dxy_serie['Loss & Profit'].kurtosis()]
skew = [dxy_serie['P&L'].skew(), dxy_serie['Loss & Profit'].skew()]
media = [dxy_serie['P&L'].mean(), dxy_serie['Loss & Profit'].mean()]
std = [dxy_serie['P&L'].std(), dxy_serie['Loss & Profit'].std()]

medidas_dp = pd.DataFrame({'Kurtosis' : kur,
                           'Sesgo' : skew,
                           'Media' : media,
                           'Vol' : std
                           },
                          index = ['P&L', 'Loss & Profit'])
medidas_dp
```
